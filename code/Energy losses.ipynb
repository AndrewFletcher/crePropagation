```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def energy(E0,a,t):
    return E0/(1+a*E0*t)
```

```python
t=np.arange(0,10,0.01)
a=1
```

```python
plt.figure(figsize=(8,8))
plt.plot(t,energy(10,a,t),'b')
plt.plot(t,energy(1,a,t),'r')
plt.show()
```
